# **Data Cleaning and Preprocessing**

```python
print(df.shape)
```

```python
# Dropping NaN values
df.dropna(inplace=True)
print(df.shape)
```

```python
df_copy = df.copy()
```

```python
df_copy.reset_index(inplace=True)
```

```python
# Splitting the dataset into train an test set
train = df_copy[df_copy['Date'] < '20150101']
test = df_copy[df_copy['Date'] > '20141231']
print('Train size: {}, Test size: {}'.format(train.shape, test.shape))
```

```python
train.columns
```

```python
# Splitting the dataset
y_train = train['Label']
train = train.iloc[:, 3:28]
y_test = test['Label']
test = test.iloc[:, 3:28]
```

```python
# Importing essential libraries for performing Natural Language Processing on given dataset
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Removing punctuation and special character from the text
train.replace(to_replace='[^a-zA-Z]', value=' ', regex=True, inplace=True)
test.replace(to_replace='[^a-zA-Z]', value=' ', regex=True, inplace=True)
```

```python
# Renaming columns
new_columns = [str(i) for i in range(0,25)]
train.columns = new_columns
test.columns = new_columns
```

```python
# Converting the entire text to lower case
for i in new_columns:
  train[i] = train[i].str.lower()
  test[i] = test[i].str.lower()
```

```python
# Joining all the columns
train_headlines = []
test_headlines = []

for row in range(0, train.shape[0]):
  train_headlines.append(' '.join(str(x) for x in train.iloc[row, 0:25]))

for row in range(0, test.shape[0]):
  test_headlines.append(' '.join(str(x) for x in test.iloc[row, 0:25]))
```

```python
train_headlines[0]
```

```python
test_headlines[0]
```

```python
# Creating corpus of train dataset
ps = PorterStemmer()
train_corpus = []

for i in range(0, len(train_headlines)):

  # Tokenizing the news-title by words
  words = train_headlines[i].split()

  # Removing the stopwords
  words = [word for word in words if word not in set(stopwords.words('english'))]

  # Stemming the words
  words = [ps.stem(word) for word in words]

  # Joining the stemmed words
  headline = ' '.join(words)

  # Building a corpus of news-title
  train_corpus.append(headline)
```

```python
# Creating corpus of test dataset
test_corpus = []

for i in range(0, len(test_headlines)):

  # Tokenizing the news-title by words
  words = test_headlines[i].split()

  # Removing the stopwords
  words = [word for word in words if word not in set(stopwords.words('english'))]

  # Stemming the words
  words = [ps.stem(word) for word in words]

  # Joining the stemmed words
  headline = ' '.join(words)

  # Building a corpus of news-title
  test_corpus.append(headline)
```

```python
# Creating the Bag of Words model
# Importing the CountVectorizer class from scikit-learn
from sklearn.feature_extraction.text import CountVectorizer

# Creating an instance of CountVectorizer
# max_features sets the maximum number of unique bigrams to consider
# ngram_range specifies the range of n-grams to generate, (2,2) means only bigrams
cv = CountVectorizer(max_features=10000, ngram_range=(2,2))

# Transforming the training corpus into a matrix of token counts
# X_train will be a NumPy array containing the counts of bigrams for each document
X_train = cv.fit_transform(train_corpus).toarray()
```

```python
X_train.shape
```

```python
#(3972, 10000) represents no of documents and no of unique words.
```

```python
X_test = cv.transform(test_corpus).toarray()
```

# **Model Building**

## *Logistic Regression*

```python
# Importing the Logistic Regression model from scikit-learn
from sklearn.linear_model import LogisticRegression

# Creating an instance of the Logistic Regression model
lr_classifier = LogisticRegression()

# Training the Logistic Regression model
# - X_train, y_train: Training data and corresponding labels
lr_classifier.fit(X_train, y_train)
```

```python
# Making predictions using the trained Logistic Regression model
# - X_test: Testing data for which predictions are made
lr_y_pred = lr_classifier.predict(X_test)
```

```python
# Accuracy, Precision and Recall
# Importing necessary metrics from scikit-learn
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Calculating accuracy, precision, and recall scores
score1 = accuracy_score(y_test, lr_y_pred)
score2 = precision_score(y_test, lr_y_pred)
score3 = recall_score(y_test, lr_y_pred)

# Printing the scores
print("---- Scores ----")
print("Accuracy score is: {}%".format(round(score1 * 100, 2)))
print("Precision score is: {}".format(round(score2, 2)))
print("Recall score is: {}".format(round(score3, 2)))
```

```python
# Importing the confusion_matrix function from scikit-learn
from sklearn.metrics import confusion_matrix

# Creating the confusion matrix
# - y_test: True class labels from the testing set
# - lr_y_pred: Predicted class labels by the Logistic Regression model
lr_cm = confusion_matrix(y_test, lr_y_pred)
```

```python
lr_cm
```

```python
# Plotting the confusion matrix
plt.figure(figsize=(10,7))
sns.heatmap(data=lr_cm, annot=True, cmap="Blues", xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Confusion Matrix for Logistic Regression Algorithm')
plt.show()
```

## *Random Forest Classifier*

```python
# Importing the RandomForestClassifier from scikit-learn
from sklearn.ensemble import RandomForestClassifier

# Creating an instance of the RandomForestClassifier
# - n_estimators sets the number of trees in the forest to 100
# - criterion='entropy' uses the information gain criterion for splitting nodes in the trees
rf_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy')

# Training the RandomForestClassifier on the training data
# - X_train: The features (input data) for training
# - y_train: The labels (output data) corresponding to the features
rf_classifier.fit(X_train, y_train)
```

```python
# Making predictions using the trained RandomForestClassifier
# - X_test: Testing data for which predictions are made
rf_y_pred = rf_classifier.predict(X_test)
```

```python
# Importing necessary functions for calculating metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Calculating accuracy, precision, and recall scores
# - y_test: Actual labels from the testing set
# - rf_y_pred: Predicted labels by the RandomForestClassifier
score1 = accuracy_score(y_test, rf_y_pred)
score2 = precision_score(y_test, rf_y_pred)
score3 = recall_score(y_test, rf_y_pred)

# Displaying the results
print("---- Scores ----")
print("Accuracy score is: {}%".format(round(score1 * 100, 2)))
print("Precision score is: {}".format(round(score2, 2)))
print("Recall score is: {}".format(round(score3, 2)))
```

```python
# Importing necessary libraries
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

# Assuming cv and lr_classifier are already defined and trained

def stock_prediction(sample_news):
    # Removing non-alphabetic characters and replacing them with spaces
    sample_news = re.sub(pattern='[^a-zA-Z]', repl=' ', string=sample_news)

    # Converting the text to lowercase
    sample_news = sample_news.lower()

    # Splitting the text into individual words
    sample_news_words = sample_news.split()

    # Removing common English stop words (e.g., "the", "and", "is")
    sample_news_words = [word for word in sample_news_words if not word in set(stopwords.words('english'))]

    # Stemming the words (reducing words to their root form)
    ps = PorterStemmer()
    final_news = [ps.stem(word) for word in sample_news_words]

    # Joining the processed words back into a string
    final_news = ' '.join(final_news)

    # Transforming the processed text into a feature vector
    temp = cv.transform([final_news]).toarray()

    # Making a prediction using the trained logistic regression classifier
    return lr_classifier.predict(temp)
```
